UNIVARIATE ANALYSIS:

Scatterplots:

```python
p = sns.barplot(x=df['Sex'], y=df['Survived'])
```

This clearly shows that more no of females survived as compared to males.

```python
p = sns.scatterplot(x=df['Age'], y=df['Fare'], hue=df['Survived'])
```

1.This shows that people who paid more fare are more likely to survive.
2.The age of the survivors varies largely between 0 and 60, and only one person above 70 years have survived. 

```python
p = sns.scatterplot(x=df['Age'], y=df['Sex'], hue=df['Survived'])
```

Again this shows more female passengers were likely to survive as green density is more in the upper part.

```python
p = sns.scatterplot(x=df['Pclass'], y=df['Fare'], hue=df['Survived'])
```

First class people, paying more ticket fare, are more likely to survive.

```python
p = sns.swarmplot(x=df['Sex'], y=df['Fare'], hue=df['Survived'])
```

CLASSIFICATION LINE PLOT:

```python
p = sns.lineplot(x =df['Pclass'], y =df['Fare'], data = df)
```

This shows that first class is more expensive than 2nd and 3rd classes.

```python
p = sns.lineplot(x =df['Pclass'], y =df['Survived'], data = df)
```
